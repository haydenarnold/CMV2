### [0720 메모] 이번 데이터(MoralBERT)의 특성 — 상관/통계분석 시 주의할 점

`Joohee/[0715]_moralBERT.ipynb`, `Joohee/predict_moralbert.py` 확인 결과:

1. **foundation별 독립 이진분류기** — `care, harm, fairness, cheating, loyalty, betrayal, authority, subversion, purity, degradation` 10개 컬럼은 각각 별도의 이진분류 모델(`vjosap/moralBERT-predict-{foundation}-in-text`)이 낸 확률입니다. 예전 SFT 방식(하나의 모델이 8개 카테고리 중 하나를 고르는 softmax, `Non-Moral`이 기준 카테고리)과 달리 **10개 컬럼이 서로 zero-sum 관계가 아니며, 뺴야 할 기준/reference 카테고리도 없습니다.**

2. **virtue-vice 쌍이 서로 양의 상관** — 이론상 반대 극성인 `care_comment`/`harm_comment`(r≈0.39), `fairness_comment`/`cheating_comment`(r≈0.44), `authority_comment`/`subversion_comment`(r≈0.38) 등이 전부 양의 상관입니다. 버그 아님 — 두 모델이 독립적으로 "이 foundation 관련 언어가 존재하는가"만 감지하기 때문에, 도덕적 수사가 많은 댓글은 두 극성의 언어를 동시에 담는 경우가 많습니다. **하나의 양극 스케일로 해석하면 안 됨.**

3. **Max-pooling으로 인한 길이 편향** — 스코어링이 문장 단위 청크의 확률 중 최댓값을 취하는 방식이라(`predict_moralbert.py`), 댓글이 길수록(청크가 많을수록) 점수가 통계적으로 올라갑니다. 실제로 `length_comment`와의 상관이 `care_comment` r≈0.35, `fairness_comment` r≈0.26 등으로 꽤 큽니다. **상관행렬을 볼 때 "도덕적 언어가 많다"가 아니라 "댓글이 길다"를 재고 있을 위험을 감안할 것.** 회귀식엔 이미 `length_comment`가 공변량으로 들어가 있어 어느 정도 통제되지만, 단순 상관(cell 9 등) 해석 시엔 주의.

4. 10개 moral score 컬럼에 결측치는 없음 (확인 완료).

### comment 변수들 logit 변환 후 다시 진행

```python
import numpy as np

eps = 1e-6

for col in comment_cols:
    x = df_merged[col].clip(lower=eps, upper=1 - eps)
    df_merged[f"{col}_logit"] = np.log(x / (1 - x))

comments_logit_cols = [f"{col}_logit" for col in comment_cols]
```

```python
challenger = df_merged[df_merged['ChallengerOP'] == 'C']
comment_cols_logit_str = ' + '.join(comments_logit_cols)
```

```python
import statsmodels.formula.api as smf

formula = f"""
    is_delta
    ~ {comment_cols_logit_str} +
    Politicality + delta_ratio + delta_ratio_c + thread_size + comment_depth + comment_score 
"""
#formula += ' + ' + ' + '.join(moral_categories)  # comment moral main effects
# + C(year)
model = smf.logit(formula, data=challenger).fit()
print(model.summary())
```
